Fixing the data type eror 

```python
# Convert "University Rank" column to numeric
df["University Rank"] = pd.to_numeric(df["University Rank"], errors="coerce")

# Filter the DataFrame based on the converted "University Rank" column
filtered_df = df[(df["University Rank"] < 100) & (df["Name of University"].notnull())]
print(filtered_df)
```
